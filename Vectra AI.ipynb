## <span style="color: green;">**Vectra Recall Data Analysis Using Pandas and Jupyter by Sharat** ðŸ˜Ž

```python
import pandas as pd
df=pd.read_csv("Demo_export.csv")

df_dataout=df[df["local_resp"]==False]
df_dataout.sort_values(by="orig_ip_bytes")


df_selected=df_dataout[["id.orig_h","orig_ip_bytes", "resp_ip_bytes", "id.resp_h"]]
df_selected_sorted=df_selected.sort_values(by="orig_ip_bytes", ascending=False)


```

## 0.2.<span style="color: orange;">ðŸ“ˆ Scatter Plot of top External IP addresses, sending data Inbound to Internal network

```python

df_selected_sorted_outbound=df_selected.sort_values(by="resp_ip_bytes", ascending=False)
df_top_outbound=df_selected_sorted_outbound.head(10)
df_top_outbound.plot(x="id.resp_h",y="resp_ip_bytes", kind="scatter", xlabel='IP Addresses', ylabel='Bytes')
```

 ##  0.3. <span style="color: orange;">ðŸ“ˆ Scatter Plot of top Internal IP addresses, sending data outbound to external network

```python
df_top=df_selected_sorted.head()
df_top.plot(x="id.orig_h",y="orig_ip_bytes", kind="scatter", xlabel='IP Addresses', ylabel='Bytes')

```

## 0.4. <span style="color: orange;">ðŸ“ˆ Filtered dataframe showing Internal and external IP talkers, sorted by inbound traffic from external IP addresses

```python
df_top_out=df_selected_sorted.sort_values(by="resp_ip_bytes", ascending=False)

df_top_out=df_top_out.head()

df_top_out
```

## 0.5.  <span style="color: orange;">ðŸ“ˆ Bar Chart Plot Of Top External Data Pullers

```python
df_dataout=df[df["local_resp"]==False]
df_dataout.sort_values(by="orig_ip_bytes")
df_dataout_new=df_dataout[["orig_ip_bytes","id.resp_h"]]
df_dataout_new.head(30).plot(kind="bar",x="id.resp_h",y="orig_ip_bytes")
```

## 0.6. <span style="color: orange;">ðŸ“ˆ Plot of Top Ports
